```python


# Write a Python function to sort a list of dictionaries by a specific key.

def sort_list_of_dicts(list_of_dicts, key, reverse=False):
    """
    Sorts a list of dictionaries by a specified key.

    Args:
        list_of_dicts (list): The list of dictionaries to sort.
        key (str): The key to sort by.
        reverse (bool, optional): If True, sort in descending order. Defaults to False.

    Returns:
        list: A new list of dictionaries sorted by the specified key.
    """
    return sorted(list_of_dicts, key=lambda x: x[key], reverse=reverse)

# Example Usage:
data = [
    {'name': 'Joseph', 'age': 26},
    {'name': 'Mwangi', 'age': 25},
    {'name': 'Kapenchun', 'age': 22}
]

print("Original list:", data)

sorted_by_age = sort_list_of_dicts(data, 'age')
print("Sorted by age (ascending):", sorted_by_age)

sorted_by_age_desc = sort_list_of_dicts(data, 'age', reverse=True)
print("Sorted by age (descending):", sorted_by_age_desc)

sorted_by_name = sort_list_of_dicts(data, 'name')
print("Sorted by name (ascending):", sorted_by_name)
```
